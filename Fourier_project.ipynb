Importing necesssary classes

```
import numpy as np
import matplotlib.pyplot as plt
import array as arr
import math
import time
```

Ideal square wave

```
l=1000
row=100
width=250
x = np.arange(1, l+1,1 )
y=np.zeros(l)
for i in range(l):
     if abs(x[i]-(l/2))<width:
         y[i]=1
plt.plot(x,y)    
```

Code for Fourier analysis

```
l=1000
row=1000
width=250
x = np.arange(1, l+1,1 )
y=[None]*l
it=np.arange(1,(row*2),2)
f=np.zeros((row,l))
for i in range(row):
    for j in range(l):
        n=((i*2)+1)
        con=(n)*(math.pi)/l
        f[i][j]=pow((-1),i)*(4/((n)*(math.pi)))*(math.sin(con*width))*(math.sin(con*(j)))
number=[1,10,20,50,80,100,200,500,1000]

for k in number:
    to_show=np.zeros(l)
    for i in range(k):
         for j in range(l):
              to_show[j]=to_show[j]+f[i][j]
    plt.plot(x,to_show)
    print(i+1," sine curves used")
    plt.show()
    #time.sleep(0.5)
    #plt.clear()
   
```

### Fourier Series Saw tooth
